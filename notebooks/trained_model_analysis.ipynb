```python
import sys
sys.path.append("..")

import os
import torch

base_path = "/workspace/dev/NumericalNet"
model_path = os.path.join(base_path, "outputs/20240526_test_v2", 'best_model.pth')
model = torch.load(model_path)
device = model.device
model.to(device)
unit_value_dict = model.unit_value_dict
```

## Digit Embedding

```python
value = [1, 2, 3, 10, 20, 30, 0.1, 0.2, 100, 1000, 20000]

embeddings = []
for v in value:
    v_ = torch.tensor([[v]], dtype=torch.float32).to(device)
    embeddings.append(model.get_digit_embedding(v_))
```

```python
VALUE = 2

input_tensor = torch.tensor([[VALUE]], dtype=torch.float32).to(device)
with torch.no_grad():
    embedding_2 = model.get_digit_embedding(input_tensor)
    embedding_2 = embedding_2.cpu()

print(embedding_2.shape)
print(embedding_2)
```

```python
units = ['cm', 'm', 'km', 'l', 'ml', 'kg', 'mg', 'g']
unit_embeddings = []
for u in units:
    u_ = torch.tensor([[unit_value_dict[u]]], dtype=torch.float32).to(device)
    with torch.no_grad():
        unit_embeddings.append(model.get_unit_embedding(u_).to('cpu'))
    
print(unit_embeddings[0].shape)
```
